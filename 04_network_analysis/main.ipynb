```python
%matplotlib inline

from datetime import datetime

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import networkx as nx

import pandas as pd
```

```python
names = ['datasource_id', # Unique id of the dataset.
         'date_donated', # Date the data was donated to the FLOSSmole project.
         'comments'] # A short textual description of the data collection.
df_datasource = pd.read_csv('data/rf_datasources.csv',
                            names=names)
```

```python
df_datasource.head(3)
```

```python
names = ['dev_loginname', # The login name of the developer working on a given project during the collection window.
         'proj_unixname', # Short name for the given project.
         'datasource_id'] # Foreign key to the datasource table. Primary key of the current table.
df_projects = pd.read_csv('data/rf_developer_projects.csv', 
                          names=names)
```

```python
len(df_projects['datasource_id'].unique())
```

```python
df_projects.head(3)
```

```python
aggregations = {'proj_unixname': 'nunique',
                'dev_loginname': 'nunique'}
aggregated_count = df_projects.groupby('datasource_id').agg(aggregations)
aggregated_count.head(3)
```

```python
merged = pd.merge(df_datasource, 
                  aggregated_count, 
                  how='left', 
                  left_on=['datasource_id'], 
                  right_on=['datasource_id'])
# dates = merged['date_donated'].map(lambda t: datetime.strptime(t, '%Y-%m-%d %H:%M:%S').year)
```

```python
# years = sorted(dates.unique())
plt.plot('date_donated', 'dev_loginname', data=merged, label='Developers')
plt.plot('date_donated', 'proj_unixname', data=merged, color='orange', label='Projects')
plt.xticks([])
# plt.xticks([datetime.strftime(datetime(year, month + 1, 1, 0, 0), '%Y-%m-%d %H:%M:%S') for year in years for month in range(12)])
plt.legend()
```

```python
g = nx.read_weighted_edgelist('data/edgelist24.csv')
degree = nx.degree(g)

g2 = g.copy()
d2 = nx.degree(g2)

for n in list(g2.nodes()):
    if d2[n] <= 1:
        g2.remove_node(n)
```

```python
g2_num_nodes = nx.number_of_nodes(g2)
g2_num_edges = nx.number_of_edges(g2)
g2_num_nodes, g2_num_edges
```

```python
d2_values = [(v) for k, v in d2]
nx.draw(g2,
        node_size=[v * 10 for v in d2_values],
        node_color='red',   # Color the node.
        edgecolors='black',
        pos=nx.spring_layout(g2))
```

```python
# To find out if the graph is fully connected.
print(nx.is_connected(g))
```

```python
# Find the most interesting graph.
graphs = list(nx.connected_component_subgraphs(g))
graphs_sorted = sorted(graphs, key=len, reverse=True)
for graph in graphs_sorted[:5]:
    num_nodes = nx.number_of_nodes(graph)
    print('num_nodes:', num_nodes)
    if num_nodes == 18:
        print('graph_degree:', nx.degree(graph))
        print()
```

```python
i = 0
for graph in graphs_sorted[:5]:
    i += 1
    print('num_nodes:', nx.number_of_nodes(graph))
    graph_degree = nx.degree(graph)
    graph_degree_values = [val for key, val in graph_degree]
    
    # Draw one set with name labels.
    f1 = plt.figure()
    nx.draw(graph,
            node_size=[v * 10 for v in graph_degree_values],
            node_color='red',   # Color the node.
            edgecolors='black',
            with_labels=True,
            font_size=8)
    filename1 = f'graphlabels{i}.png'
    f1.savefig(filename1)
    
    # Draw one set without name labels.
    f2 = plt.figure()
    nx.draw(graph,
            node_size=[v * 10 for v in graph_degree_values],
            node_color='red',   # Color the node.
            edgecolors='black')
    filename2 = f'graph{i}.png'
    f2.savefig(filename2)
```

```python
i = 0
for graph in graphs_sorted[:5]:
    i += 1
    print(nx.number_of_nodes(graph))
    graph_degree = nx.degree(graph)
    
    # Find cliques.
    cliques = list(nx.find_cliques(graph))
    print('cliques for graph', i)
    print(cliques)
    
    # Calculate eigenvector centrality.
    ev = nx.eigenvector_centrality_numpy(graph)
    ev_sorted = sorted(ev.items(), 
                       key=lambda t: t[1], 
                       reverse=True)
    for key, val in ev_sorted:
        print(key, val)
```

```python
G = nx.read_weighted_edgelist('data/edgelist64.csv')
graph_degree = nx.degree(G)
nx.draw(G,
        pos=nx.spring_layout(G),
        node_size=[v * 10 for k, v in graph_degree],
        node_color='red',
        edgecolors='black',
        with_labels=False,
        font_size=8)

node_list = ['tirsen', 'shen', 'mlee', 'ged', 'objo', 'stellsmi', 'cowboyd', 'asong', 'christkv', 'hisnice', 'duelin_markers', 'stillflame']
nx.draw_networkx_nodes(G,
                       pos=nx.spring_layout(G),
                       nodelist=node_list,
                       node_color='green',
                       edgecolors='black')
```

```python
G = nx.read_weighted_edgelist('data/edgelist64.csv')
graphs = list(nx.connected_component_subgraphs(G))
conncomp = graphs[0]

ego = nx.Graph(nx.ego_graph(conncomp, 'tirsen', radius=3))
graph_degree = nx.degree(ego)
pos = nx.spring_layout(ego)
nx.draw(ego,
        pos,
        node_size=[v * 10 for k, v in graph_degree],
        with_labels=True,
        font_size=8)
nx.draw_networkx_nodes(ego,
                       pos,
                       nodelist=['tirsen'],
                       node_size=300, 
                       node_color='g')
```
